````c++
//MBCS : UTF-8  
#include <stdio.h>
#include <locale.h>

int main() {
    setlocale(LC_ALL, "");

    char* str = "";

    printf(": %s\n", str);
    printf("UTF-8 : ");
    for (int i = 0; str[i] != '\0'; i++) {
        printf("%02X ", (unsigned char)str[i]);  // 16 : %02X
    }
    printf("\n");

    return 0;
}
````
