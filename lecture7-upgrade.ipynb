````c++
//MBCS : UTF-8 바이트 확인
#include <stdio.h>
#include <locale.h>

int main() {
    setlocale(LC_ALL, "");

    char* str = "한";

    printf("문자: %s\n", str);
    printf("UTF-8 바이트: ");
    for (int i = 0; str[i] != '\0'; i++) {
        printf("%02X ", (unsigned char)str[i]);  // 16진법으로 표현: %02X
    }
    printf("\n");

    return 0;
}
````
