# [숙제] LAB 5-2 if문을 switch문으로 전환하기

### if --> switch 전환

```c
#include <stdio.h>
int main(){
    int n;
    int chk;
    printf("정수를 입력하세요? ");
    scanf("%d",&n);

    chk = !(n%2) + 2*!(n%3);
    switch(chk){
        case 1:
            printf("2의 배수: %d(%d)\n", n, chk);
            break;
        case 2:
            printf("3의 배수: %d(%d)\n", n, chk);
            break;
        case 3:
            printf("2의 배수이면서 3의 배수: %d(%d)\n", n, chk);
            break;
        default:
            printf("기타 숫자: %d(%d)\n", n, chk);
            break;
    }
}
```

# 사칙연산을 switch문으로 완성하시오.
### 입력값  (두개의 정수값과 연산자: +, -, *, /)
> 아래와 같이 입력하시오.  
> 2+3
### 출력값
> 2 + 3 = 5


````c++
//공백을 이용해서 입력받기 과제 2
#include <stdio.h>
int main(){
    printf("아래와 같이 입력하시오\r\n.");
    int a, b;
    char c;
    scanf("%d %c %d",&a,&c,&b);
    switch(c){
    case '+': printf("%d + %d = %d\n", a, b, a+b); break;
    case '-': printf("%d - %d = %d\n", a, b, a-b); break;
    case '*': printf("%d * %d = %d\n", a, b, a*b); break;
    case '/': printf("%d // %d = %d\n", a, b, a/b); break;
    
    }    
}
````
