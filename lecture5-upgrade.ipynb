```c
#include <stdio.h>
int main(){
    int n;
    int chk;
    printf(" ? ");
    scanf("%d",&n);

    chk = !(n%2) + 2*!(n%3);
    switch(chk){
        case 1:
            printf("2 : %d(%d)\n", n, chk);
            break;
        case 2:
            printf("3 : %d(%d)\n", n, chk);
            break;
        case 3:
            printf("2  3 : %d(%d)\n", n, chk);
            break;
        default:
            printf(" : %d(%d)\n", n, chk);
            break;
    }
}
```

````c++
//    2
#include <stdio.h>
int main(){
    printf("  \r\n.");
    int a, b;
    char c;
    scanf("%d %c %d",&a,&c,&b);
    switch(c){
    case '+': printf("%d + %d = %d\n", a, b, a+b); break;
    case '-': printf("%d - %d = %d\n", a, b, a-b); break;
    case '*': printf("%d * %d = %d\n", a, b, a*b); break;
    case '/': printf("%d // %d = %d\n", a, b, a/b); break;
    
    }    
}
````
