````c++
#include <stdio.h>
int main(){
    int year; //= 2025;
    int month; //= 3;
    int day; //= 17;
    char gender[5]; //= "성별";
    char name[20]; //= "김나라";

    printf("태어난 연도, 월, 일, 성별, 이름을 공백으로 구분하여 입력하세요. \n");
    scanf("%d%d%d%s%s",&year, &month, &day, gender, name); // & 단일변수는 단일 주소 사용 , 배열은 이름이 주소

    printf("연도: %d \n 월: %d \n 일: %d \n 성별: %s \n 이름: %s", year, month, day, gender, name);
}
````

## [코드 2-13] 메아리 프로그램

```c
#include <stdio.h>

int main(){
    char input[128] = { 0 };

    printf("입력> ");
    scanf("%[^\n]s", input);

    printf("출력> %s", input);
}
```
<hr>  
입력&gt; <font color=red> 난생처음 C 프로그래밍</font><br>
출력&gt; 난생처음 C 프로그래밍<br>

````c++
#include <stdio.h>

int main(){
    char input[128] = { 0 };

    printf("입력> ");
    
    scanf("%[^\n]s", input); // [^(문자의 처음)부터 \n을 만날 때까지 (enter를 누를 때 까지) 하나의 문자로 인식한다] []는 범위의 역할

    printf("출력> %s", input);
}
````
