    int result;

    for(int i=1;i<=100;i++){

        }
    }
}

````c++
#include <stdio.h>
int main(){
    int result;
    int count = 0;
    for(int i=1;i<=100;i++){
        if(i%5 == 0){
            printf("5 [%d] -> %d\n",++count,i);
            if(count==17) {
                result = i;
                break;
            }
        }
    }
    printf("17 5  : %d", result);
}
````

````c++
#include <stdio.h>
int main(){
    int dif = 'a'-'A';
    for(int i='A';i<='z';i++){
        if(i<='Z'){// >> 
            printf("%c : %d\n",i+dif,i+dif);}//A ~ Z
        else if(i>='a'){//  >> 
            printf("%c : %d\n",i-dif,i-dif);}
        else{
            printf("%c : %d\n",i,i);}//
    }
    return 0;
}
````

````c++
// (sum1), (sum2) 
#include <stdio.h>
int main(){
    int i=0,sum=0;
    int sum1=0,sum2=0;
    while(1){
        sum = sum + (i+1);        
        if(i%2==0) sum1 = sum1 + (i+1);
        if(i%2) sum2 = sum2 + (i+1);
        i++;
        
        if(i>99) break;        
    }
    printf("sum1 = %d\n",sum1);
    printf("sum2 = %d\n",sum2);
    printf("sum = %d\n",sum);
}
````
