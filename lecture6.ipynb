### (1 ~ 100까지 정수값을 순차적으로 증가시킨다고 가정할 때, 5의 배수중 17번째 정수값을 구하고, 순환문을 중지하는 프로그램을 작성하시오.
결과
>>> 5배수 [1] -> 5  
>>> 5배수 [2] -> 10  
>>> 5배수 [3] -> 15  
>>> ...  
>>> 5배수 [17] -> 85 ***

<div style="font-size:17px;">code<br>
<pre style="color:white">
#include <stdio.h>
int main(){
    int result;
    int count = 0 ;
    for(int i=1;i<=100;i++){
        if(!(i%5)){
            printf("5배수 [%d] -> %d\n",++count,i);
            if(count==17) break;
        }
    }
}    
</pre>
</div>

````c++
#include <stdio.h>
int main(){
    int result;
    int count = 0;
    for(int i=1;i<=100;i++){
        if(i%5 == 0){
            printf("5배수 [%d] -> %d\n",++count,i);
            if(count==17) {
                result = i;
                break;
            }
        }
    }
    printf("17번째 5의 배수 : %d", result);
}
````

````c++
#include <stdio.h>
int main(){
    int sum = 0;

    for(int i=0; i<100; i++){
        sum = sum + (i + 1);
    }
    printf("총합: %d", sum);
}// while >> for문
````

````c++
// 홀수합(sum1), 짝수합(sum2) 추가하기
#include <stdio.h>
int main(){
    int i=0,sum=0;
    int sum1=0,sum2=0;
    while(1){
        sum = sum + (i+1);        
        if(i%2==0) sum1 = sum1 + (i+1);
        if(i%2) sum2 = sum2 + (i+1);
        i++;
        
        if(i>99) break;        
    }
    printf("sum1 = %d\n",sum1);
    printf("sum2 = %d\n",sum2);
    printf("sum = %d\n",sum);
}
````
